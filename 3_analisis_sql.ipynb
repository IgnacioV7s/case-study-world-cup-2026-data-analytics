# Caso de Estudio: Fase de Análisis con SQL (Analyze)
## Proyecto: Base de Datos Relacional del Mundial 2026

Para demostrar competencias avanzadas en el manejo de bases de datos, cargaremos el dataset limpio en un motor **SQLite**. Realizaremos consultas e hipótesis de negocio utilizando sintaxis SQL estándar para responder preguntas demográficas y de distribución global de los jugadores.

```python
import sqlite3
import pandas as pd

# 1. Conectar a una base de datos SQLite (se creará un archivo llamado mundial2026.db)
conn = sqlite3.connect("mundial2026.db")

# 2. Leer el CSV limpio que guardamos en el paso anterior
df_limpio = pd.read_csv("plantillas_limpias_mundial2026.csv")

# 3. Guardar el DataFrame como una tabla SQL llamada 'jugadores'
df_limpio.to_sql("jugadores", conn, if_exists="replace", index=False)

print("¡Éxito! Base de Datos 'mundial2026.db' creada y tabla 'jugadores' cargada.")
```

### Consulta 1: Top 10 de clubes que más jugadores aportan al Mundial
Esta consulta identifica qué instituciones de fútbol tienen mayor impacto en el torneo, ordenadas de mayor a menor.

```python
query_clubes = """
SELECT Club, COUNT(*) AS Total_Jugadores
FROM jugadores
GROUP BY Club
ORDER BY Total_Jugadores DESC
LIMIT 10;
"""

# Ejecutar la consulta y mostrar el resultado en una tabla limpia
pd.read_sql_query(query_clubes, conn)
```

### Consulta 2: Perfil Demográfico por Posición
Para entender la madurez táctica del torneo, analizaremos la edad promedio, la edad mínima y la máxima agrupadas por cada posición en el campo de juego. Esto nos revelará qué posiciones tienden a requerir mayor experiencia.

```python
query_edades = """
SELECT 
    Posicion, 
    COUNT(*) AS Total_Jugadores,
    ROUND(AVG(Edad), 1) AS Edad_Promedio,
    MIN(Edad) AS Edad_Minima,
    MAX(Edad) AS Edad_Maxima
FROM jugadores
GROUP BY Posicion
ORDER BY Edad_Promedio DESC;
"""

pd.read_sql_query(query_edades, conn)
```

### Consulta 3: Clubes con Alta Concentración de Convocados (Filtro HAVING)
A diferencia de un `WHERE` que filtra filas individuales, utilizaremos `HAVING` para filtrar grupos de datos. En este caso, aislaremos únicamente a los clubes de Élite que aportan 5 o más jugadores al Mundial.

```python
query_elite = """
SELECT Club, COUNT(*) AS Convocados
FROM jugadores
GROUP BY Club
HAVING Convocados >= 5
ORDER BY Convocados DESC;
"""

pd.read_sql_query(query_elite, conn)
```

### Buenas Prácticas en Ingeniería de Datos
Una vez finalizado el análisis, es fundamental cerrar la conexión a la base de datos para liberar los recursos del sistema y evitar corrupción de archivos en el disco.

```python
# Cerrar la conexión de manera segura
conn.close()
print("Conexión a 'mundial2026.db' cerrada correctamente. Datos asegurados.")
```

### 1. 🎯 Madurez Táctica según la Posición en el Campo
Los datos reflejan una clara correlación entre la posición de los futbolistas y su promedio de edad, dividiendo las prioridades de los cuerpos técnicos en dos enfoques:
* **La Experiencia en la Retaguardia:** Los Arqueros (`GK`) y Defensas (`DF`) registran los promedios de edad más altos del torneo (rango de 27.5 a 29.8 años). Las selecciones priorizan el liderazgo mental, la toma de decisiones bajo presión y el posicionamiento táctico para proteger su propio arco.
* **La Juventud y Explosividad en el Ataque:** Los Mediocampistas (`MF`) y Delanteros (`FW`) presentan promedios de edad significativamente menores. El fútbol moderno exige un despliegue físico intenso, velocidad en transiciones y resistencia para la presión alta, características que los entrenadores buscan en perfiles más jóvenes.

### 2. 🏛️ Centralización del Talento en la Élite Global
La consulta relacional avanzada con el filtro `HAVING` demostró que el talento del Mundial no está distribuido equitativamente, sino concentrado en un oligopolio de clubes:
* Un grupo selecto de instituciones pertenecientes a las 5 grandes ligas europeas (Premier League, LaLiga, Serie A, Bundesliga, Ligue 1) y ligas de alto poder económico concentran el mayor volumen de jugadores mundialistas.
* **Conclusión de Negocio:** El éxito deportivo en el torneo está directamente ligado al nivel de competencia semanal. Las selecciones con mayor cantidad de jugadores compitiendo bajo el estándar de la UEFA Champions League y ligas de élite cuentan con una ventaja competitiva crítica en ritmo de juego.

### 3. 🌍 Modelos de Gestión de Plantilla: Locales vs. Globales
Al evaluar la procedencia de los convocados por liga de origen, se observan dos estrategias drásticamente opuestas para armar un equipo competitivo:
* **Modelos de Arraigo Local:** Potencias europeas (como Inglaterra o España) presentan plantillas donde casi el 100% de sus jugadores militan en sus ligas domésticas. Esto beneficia la cohesión, la identidad táctica y disminuye el desgaste por viajes.
* **Modelos de Exportación Global:** Selecciones de Latinoamérica y África dependen casi en su totalidad (más del 85%) de jugadores que militan en el extranjero. Su fortaleza radica en el roce internacional y la diversidad de estilos que sus atletas absorben en clubes de todo el mundo.
